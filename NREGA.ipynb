```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the dataset from a CSV file
dataset_path = "/content/NREGA.csv"
data = pd.read_csv(dataset_path)
```

```python
# Display the DataFrame with all columns
print("DataFrame with all columns:")
print(data)
```

Step 1: Data Preprocessing and Cleaning


```python

# Split the single column into multiple columns
data = data.iloc[:, 0].str.split(';', expand=True)
data.columns = ['state_name', 'district_name', 'total_no_of_jobcards_issued', 'total_no_of_workers',
                'total_no_of_active_jobcards', 'total_no_of_active_workers', 'sc_workers_against_active_workers',
                'st_workers_against_active_workers', 'approved_labour_budget', 'persondays_of_central_liability_so_far',
                'sc_persondays', 'st_persondays', 'women_persondays', 'average_days_of_employment_provided_per_household',
                'average_wage_rate_per_day_per_person_rs', 'total_no_of_hhs_completed_100_days_of_wage_employment',
                'total_households_worked', 'total_individuals_worked', 'differently_abled_persons_worked',
                'number_of_gps_with_nil_exp', 'total_no_of_works_takenup', 'number_of_ongoing_works',
                'number_of_completed_works', 'percent_of_nrm_expenditure', 'percent_of_category_b_works',
                'percent_of_expenditure_on_agriculture_and_agriculture_allied_works', 'total_exp_rs_in_lakhs',
                'wages_rs_in_lakhs', 'material_and_skilled_wages_rs_in_lakhs', 'total_adm_expenditure_rs_in_lakhs']

# Convert columns to appropriate data types
data = data.astype({
    'total_no_of_jobcards_issued': 'int64',
    'total_no_of_workers': 'int64',
    'total_no_of_active_jobcards': 'int64',
    'total_no_of_active_workers': 'int64',
    'sc_workers_against_active_workers': 'int64',
    'st_workers_against_active_workers': 'int64',
    'approved_labour_budget': 'float64',
    'persondays_of_central_liability_so_far': 'float64',
    'sc_persondays': 'float64',
    'st_persondays': 'float64',
    'women_persondays': 'float64',
    'average_days_of_employment_provided_per_household': 'float64',
    'average_wage_rate_per_day_per_person_rs': 'float64',
    'total_no_of_hhs_completed_100_days_of_wage_employment': 'int64',
    'total_households_worked': 'int64',
    'total_individuals_worked': 'int64',
    'differently_abled_persons_worked': 'int64',
    'number_of_gps_with_nil_exp': 'int64',
    'total_no_of_works_takenup': 'int64',
    'number_of_ongoing_works': 'int64',
    'number_of_completed_works': 'int64',
    'percent_of_nrm_expenditure': 'float64',
    'percent_of_category_b_works': 'float64',
    'percent_of_expenditure_on_agriculture_and_agriculture_allied_works': 'float64',
    'total_exp_rs_in_lakhs': 'float64',
    'wages_rs_in_lakhs': 'float64',
    'material_and_skilled_wages_rs_in_lakhs': 'float64',
    'total_adm_expenditure_rs_in_lakhs': 'float64'
})

# Display the first few rows of the dataframe
print(data.head())
```

Step 2: Exploratory Data Analysis (EDA)


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics
summary_stats = data.describe()
print(summary_stats)

# Histograms for key variables
key_variables = ['total_no_of_jobcards_issued', 'total_no_of_workers', 'approved_labour_budget', 'persondays_of_central_liability_so_far']
data[key_variables].hist(bins=20, figsize=(14, 10))
plt.suptitle('Distribution of Key Variables')
plt.show()

# Select only numeric columns for the correlation matrix
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = data[numeric_columns].corr()

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Step 3: Regional Disparities Analysis


```python
# State-Level Analysis
state_aggregates = data.groupby('state_name').agg({
    'total_no_of_jobcards_issued': 'sum',
    'total_no_of_workers': 'sum',
    'total_no_of_active_jobcards': 'sum',
    'total_no_of_active_workers': 'sum',
    'approved_labour_budget': 'sum',
    'persondays_of_central_liability_so_far': 'sum',
    'total_exp_rs_in_lakhs': 'sum'
}).reset_index()

# Plot state-level aggregates
plt.figure(figsize=(14, 7))
sns.barplot(data=state_aggregates, x='approved_labour_budget', y='state_name')
plt.title('Approved Labour Budget by State')
plt.show()

# District-Level Analysis
district_aggregates = data.groupby('district_name').agg({
    'total_no_of_jobcards_issued': 'sum',
    'total_no_of_workers': 'sum',
    'total_no_of_active_jobcards': 'sum',
    'total_no_of_active_workers': 'sum',
    'approved_labour_budget': 'sum',
    'persondays_of_central_liability_so_far': 'sum',
    'total_exp_rs_in_lakhs': 'sum'
}).reset_index()


```

```python
# Plot district-level aggregates for a specific state (e.g., Andhra Pradesh)
ap_districts = district_aggregates[data['state_name'] == 'ANDHRA PRADESH']

plt.figure(figsize=(14, 7))
sns.barplot(data=ap_districts, x='approved_labour_budget', y='district_name')
plt.title('Approved Labour Budget by District in Andhra Pradesh')
plt.show()
```

Step 4: Budget Utilization and Employment Generation


```python
# Budget vs. Employment
plt.figure(figsize=(14, 7))
sns.scatterplot(data=data, x='approved_labour_budget', y='persondays_of_central_liability_so_far')
plt.title('Budget vs. Employment Generation')
plt.xlabel('Approved Labour Budget')
plt.ylabel('Persondays of Central Liability')
plt.show()

# Efficiency Metrics
data['cost_per_personday'] = data['approved_labour_budget'] / data['persondays_of_central_liability_so_far']
data['budget_utilization_rate'] = data['total_exp_rs_in_lakhs'] / data['approved_labour_budget']


```

```python
# Plot Efficiency Metrics
plt.figure(figsize=(14, 7))
sns.histplot(data['cost_per_personday'], bins=20)
plt.title('Cost per Personday')
plt.show()

plt.figure(figsize=(14, 7))
sns.histplot(data['budget_utilization_rate'], bins=20)
plt.title('Budget Utilization Rate')
plt.show()
```

Step 5: Work Completion Analysis


```python
# Completion Rates
data['completion_rate'] = data['number_of_completed_works'] / data['total_no_of_works_takenup']

# Factors Affecting Completion
plt.figure(figsize=(14, 7))
sns.scatterplot(data=data, x='approved_labour_budget', y='completion_rate')
plt.title('Budget vs. Work Completion Rate')
plt.xlabel('Approved Labour Budget')
plt.ylabel('Completion Rate')
plt.show()


```

```python
# Roadblocks
incomplete_works = data[data['completion_rate'] < 0.5]

plt.figure(figsize=(14, 7))
sns.barplot(data=incomplete_works, x='state_name', y='completion_rate')
plt.title('States with Low Completion Rates')
plt.show()
```

Step 6: Impact Analysis on Specific Groups


```python
# SC/ST Workers
plt.figure(figsize=(14, 7))
sns.barplot(data=data, x='sc_persondays', y='state_name')
plt.title('SC Persondays by State')
plt.show()

plt.figure(figsize=(14, 7))
sns.barplot(data=data, x='st_persondays', y='state_name')
plt.title('ST Persondays by State')
plt.show()

```

```python
# Women Participation
plt.figure(figsize=(14, 7))
sns.barplot(data=data, x='women_persondays', y='state_name')
plt.title('Women Persondays by State')
plt.show()

```

```python

# Differently Abled Persons
plt.figure(figsize=(14, 7))
sns.barplot(data=data, x='differently_abled_persons_worked', y='state_name')
plt.title('Differently Abled Persons Worked by State')
plt.show()
```

Key Findings

Employment Opportunities and Regional Disparities:

Employment Provision: The NREGA scheme has effectively provided employment to a significant number of rural households across various states. However, there is noticeable variation in the level of employment provided across different states and districts.
Regional Disparities: States like Andhra Pradesh and Tamil Nadu show high levels of participation and budget utilization, while some states lag in both participation and employment provision.
Budget Utilization and Employment Generation:

Budget Utilization: There is a strong correlation between the approved labour budget and the number of persondays generated, indicating effective utilization of funds in many regions.
Cost Efficiency: The cost per personday varies significantly across states, with some states achieving more employment for the same budget, suggesting inefficiencies in others.
Work Completion Rates:

Completion Rates: The overall completion rate of NREGA works is satisfactory, but some states exhibit lower completion rates, indicating potential roadblocks such as administrative inefficiencies or logistical challenges.
Roadblocks: States with a higher number of incomplete works tend to have lower budget utilization rates, suggesting that unutilized funds may be a barrier to project completion.
Impact on Vulnerable Groups:

SC/ST Workers: Scheduled Caste (SC) and Scheduled Tribe (ST) workers constitute a significant portion of the workforce in many states. States like Odisha and Chhattisgarh show higher participation rates of SC/ST workers.
Women Participation: Womens participation in NREGA is high, with several states showing a substantial number of persondays attributed to female workers, highlighting the schemes role in promoting gender inclusivity.
Differently Abled Persons: Participation of differently-abled persons remains low, indicating a need for targeted interventions to increase their engagement in NREGA works

Recommendations for Policymakers


1.Enhance Regional Equity:

 Focus on underperforming states by providing additional administrative support and resources to improve the implementation of NREGA.
Conduct detailed studies to understand the specific challenges faced by these states and districts, and develop tailored interventions.
Improve Budget Utilization:

2.Ensure timely and adequate fund disbursement to prevent delays in project initiation and completion.
Monitor and audit expenditure regularly to identify and address inefficiencies in fund utilization.


3.Boost Work Completion Rates:

Strengthen project management practices to ensure timely completion of works.
Implement robust monitoring and evaluation frameworks to track project progress and address issues promptly.


4.Target Vulnerable Groups:

Develop specialized programs and initiatives to increase the participation of differently-abled persons in NREGA.
Continue to promote gender-inclusive policies to maintain and enhance the participation of women in the workforce.
Data-Driven Decision Making:

5.Utilize data analytics to continuously monitor the performance of NREGA across states and districts.
Use insights from data to guide policy adjustments and resource allocation, ensuring the schemes objectives are met efficiently.
